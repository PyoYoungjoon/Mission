```python
import pandas as pd 

pileup_Mirlet7g = pd.read_csv('CLIP-let7g-gene.pileup', sep = '\t', names = ['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup_Mirlet7g.tail()

pileup_Mirlet7f_1 = pd.read_csv('CLIP-let7f_1-gene.pileup', sep = '\t', names = ['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup_Mirlet7f_1.tail()

pileup_Mirlet7d = pd.read_csv('CLIP-let7d-gene.pileup', sep = '\t', names = ['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup_Mirlet7d.tail()
```

```python
import re
toremove = re.compile('[<>$*#^]')
pileup_Mirlet7g['matches'] = pileup_Mirlet7g['basereads'].apply(lambda x: toremove.sub('',x))
pileup_Mirlet7f_1['matches'] = pileup_Mirlet7f_1['basereads'].apply(lambda x: toremove.sub('',x))
pileup_Mirlet7d['matches'] = pileup_Mirlet7d['basereads'].apply(lambda x: toremove.sub('',x))
```

```python
import pandas as pd
import numpy as np
from collections import Counter

bed_Mirlet7g= pileup_Mirlet7g[['chrom', 'pos']].copy()
bed_Mirlet7g.columns = ['chr', 'start_position']
bed_Mirlet7g['end_position'] = bed_Mirlet7g['start_position']+1



def calculate_entropy(row):
    bases = ['A', 'T', 'C', 'G']
    entropy = 0
    count_dict = Counter(row)
    for base in bases:
        count = count_dict.get(base, 0)
        total = sum(count_dict.values())
        p = count / total if total > 0 else 0
        if p > 0:
            entropy -= p * np.log2(p)
    return entropy

bed_Mirlet7g['entropy'] = pileup_Mirlet7g['matches'].str.upper().apply(calculate_entropy).copy()
bed_Mirlet7g.to_csv('Mirlet7g.bedgraph', sep='\t', header=False, index=False)
bed_Mirlet7g
```
